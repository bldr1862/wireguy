```python
from pathlib import Path
from utils import io as IO
from PIL import Image
import torch
import pandas as pd
import cv2
from tqdm import tqdm
```

## Overlay images and GT to understand labels

```python
# you can skip this cell
run_overlay = False
if run_overlay:
    overlay_dir = "data/overlay"
    Path(overlay_dir).mkdir(parents=True, exist_ok=True)
    for i, row in tqdm(consolidation.iterrows(), total=len(consolidation)):
        image_path = row["image_path"]
        label_path = row["label_path"]

        image = IO.load_image(image_path)
        mask = IO.load_image(label_path)

        overlay = cv2.addWeighted(image,0.9,mask,0.1,0)
        dst_path = Path(overlay_dir).joinpath(f"{row['stem']}.png")
        # the channels are shifted but it's enough to understand the data
        cv2.imwrite(str(dst_path), cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))
    
```

## Overlay image preds

```python
# you can skip this cell
path = "D:/projects/ikeGPS/results/168584611301"
run_overlay = True
if run_overlay:
    overlay_train_dir = Path(path).joinpath("overlay/train")
    overlay_val_dir = Path(path).joinpath("overlay/val")
    overlay_train_dir.mkdir(exist_ok=True, parents=True)
    overlay_val_dir.mkdir(exist_ok=True, parents=True)
    for mask_path in Path(path + "/train").glob("*.png"):
        print(mask_path)
        img_path = Path("D:\projects\ikeGPS\data\ISG-Guywire_Ground_Seg").joinpath(f"{mask_path.stem}.JPG")
        mask = IO.load_image(mask_path)
        img = IO.load_image(img_path, resize=(644,644))
        overlay = cv2.addWeighted(img,0.3,mask,0.7,0)
        cv2.imwrite(str(overlay_train_dir.joinpath(f"{mask_path.stem}.png")), cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))

    for mask_path in Path(path + "/val").glob("*.png"):
        print(mask_path)
        img_path = Path("D:\projects\ikeGPS\data\ISG-Guywire_Ground_Seg").joinpath(f"{mask_path.stem}.JPG")
        mask = IO.load_image(mask_path)
        img = IO.load_image(img_path, resize=(644,644))
        overlay = cv2.addWeighted(img,0.3,mask,0.7,0)
        cv2.imwrite(str(overlay_val_dir.joinpath(f"{mask_path.stem}.png")), cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))
```
